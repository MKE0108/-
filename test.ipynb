```python
import cv2
import numpy as np

def mse(imageA, imageB):
    # MSE 計算公式是兩個圖像的差的平方，然後求平均值
    err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
    err /= float(imageA.shape[0] * imageA.shape[1])
    return err
```

```python
import cv2
from skimage import io
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt
import cv2
from windows import *
# 讀取圖像
y=130/1080
x=550/1920
w=820/1920
h=840/1080
img1 = cv2.imread('task2_auto_ac/ac_1.png')
img1=img1[int(y*img1.shape[0]):int((y+h)*img1.shape[0]),int(x*img1.shape[1]):int((x+w)*img1.shape[1])]
img1=cv2.resize(img1,(100,100))




# 將圖像轉換為灰度圖
# img2 = capture_window("League of Legends")
img2 = cv2.imread('task2_auto_ac/waiting1.png')
img2=img2[int(y*img2.shape[0]):int((y+h)*img2.shape[0]),int(x*img2.shape[1]):int((x+w)*img2.shape[1])]
img2=cv2.resize(img2,(100,100))

# 計算 SSIM
ssim_index, ssim_map = ssim(img1, img2, full=True)
print(f"SSIM: {ssim_index}")
if(ssim_index>0.9):
    print("found")
    

```
